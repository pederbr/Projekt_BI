```python
import pandas as pd

def convert_to_statistikk_alle(input_file, output_file):
    f = open(output_file, "w") 

    df = pd.read_csv(input_file)

    with open(output_file, 'w') as file:
        file.write('')
    
    q_list=[
    df.loc[72:76],  #fremforing_presentasjon
    df.loc[65:69],  #innhold_presentasjon
    df.loc[79:83],  #kunnskap_bedrift_pre
    df.loc[86:90],  #kunnskap_bedrift_post
    df.loc[93:97],  #info_jobb
    df.loc[58:62],  #mingling
    df.loc[17:21],  #intterresant_arbeid
    df.loc[33:37],  #sosialt_miljø
    df.loc[44:48],  #arbeidsvilkår
    df.loc[51:55]]  #helhetsvurdering
    


    def init_file():
        init_dict={"5":[],"4":[],"3":[],"2":[],"1":[]}
        for elem in q_list:
            list_from_var=elem.values.tolist()
            for i in range(5):
                value=int(list_from_var[i][1])
                key=str(5-(i))
                init_dict[key].append(value)
            

        df_out=pd.DataFrame(init_dict)
        df_out.to_csv(output_file, mode='a', index=False) 


    def append_to_file(dataset):
        t_dict={}
        t_list=dataset.values.tolist()
        for elem in t_list:
            t_dict[elem[0]]=[int(elem[1])]
        df_out=pd.DataFrame(t_dict)
        df_out.to_csv(output_file, mode='a', index=False)
    
    
    init_file()
    
    append_to_file(df.loc[40:41])

    studie=df.loc[6:7]
    studie_dict={}
    studie_list = studie.values.tolist()
    studie_dict["data"]=[int(studie_list[0][1])]
    studie_dict["komtek"]=[int(studie_list[1][1])]

    df_out=pd.DataFrame(studie_dict)
    #print(df_out)
    df_out.to_csv(output_file, mode='a', index=False)

    append_to_file(df.loc[1:3])
    append_to_file(df.loc[10:14])

    f.close()

convert_to_statistikk_alle("WS_e_tjenesten.csv", "test1.csv")

```
